```python
# 20221003T050000.000 GMT
query = (
    df.filter(pl.col("pyIsPropositionActive") != "Never")
    .with_columns(
        pl.col('StartDate', 'EndDate')
        .str.replace(" GMT", "")
        .str.strptime(pl.Datetime, fmt="%Y%m%dT%H%M%S%.f"),
        pl.col("pxUpdateDateTime")
        .cast(pl.Datetime).dt.with_time_unit("ms")
        .alias("pxUpdateDateTime_")
    )
    .select('StartDate', 'EndDate', 'pxUpdateDateTime_', pl.col('^py.*$'))
)
query
# StartDate
#EndDate
```
